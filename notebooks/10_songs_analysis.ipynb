```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_audio_features = pd.read_csv("1_spotify_10_songs.csv")
df_audio_features.columns = df_audio_features.columns.str.strip()
df_audio_features.set_index(['song_name', 'artist'], inplace=True)
df_audio_features.drop(['id', 'html'], axis=1, inplace=True)
df_audio_features
```

## Min Max Scaler

```python
# Import the transformer 
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize parameters (optionally set the parameters)
my_min_max = MinMaxScaler(feature_range=(0,1))
```

```python
# Fit the transformer to the data
my_min_max.fit(df_audio_features)
```

```python
# Use the transformer to transform the data 
scaled_audio_features = my_min_max.transform(df_audio_features)
```

```python
# Reconvert transformed data into a dataframe 
min_max_scaled_df= pd.DataFrame(scaled_audio_features,
            index=df_audio_features.index,
            columns=df_audio_features.columns)
```

```python
# Display 'energy' column of original dataframe and scaled dataframe
pd.DataFrame({'original_energy_column' : df_audio_features.iloc[:,1],
              'scaled_energy_column' : min_max_scaled_df.iloc[:,1]}
            ).sort_values(by='original_energy_column')
```

## Standart Scaler

```python
# Import the transformer 
from sklearn.preprocessing import StandardScaler

# Initialize paramteres
my_standart_scaler = StandardScaler()

# Fit the transformer to the data
my_standart_scaler.fit_transform(df_audio_features)

# Use the transformer to transform the data
standart_audio_features = my_standart_scaler.transform(df_audio_features)

# Convert transformed data into a dataframe
standart_df = pd.DataFrame(standart_audio_features, 
                           columns= df_audio_features.columns,
                           index=df_audio_features.index)

# Display 'energy' column of original dataframe and scaled dataframe
pd.DataFrame({'original_energy_column': df_audio_features.iloc[:,1],
             'standart_scaled_column' : standart_df.iloc[:,1]}
            ).sort_values(
                        by='original_energy_column')
```

## Robust Scaler

```python
# Import the transformer
from sklearn.preprocessing import RobustScaler

# Initialize the parameters
my_robust_scaler = RobustScaler()

# Fit the transformer to the data
my_robust_scaler.fit_transform(df_audio_features)

# Use the transformer to transform the data
robust_audio_features = my_robust_scaler.transform(df_audio_features)

# Convert transformed data into a dataframe
robust_scaled_df = pd.DataFrame(robust_audio_features, 
                               columns= df_audio_features.columns,
                               index= df_audio_features.index)

# Display 'energy' column of original dataframe and scaled dataframe
pd.DataFrame({'original_energy_column': df_audio_features.iloc[:,1],
             'robust_scaled_column' : robust_scaled_df.iloc[:,1]}
            ).sort_values(
              by='original_energy_column')
```

# Quantile Transformer

## What Does QuantileTransformer Do?
From a programming and ML perspective:
*It ranks the values of each feature (like sorting them).
*Then, it maps these ranks to a uniform (or normal) distribution.
*This helps make features more comparable, especially if they're skewed or have outliers.
*Why use it instead of MinMaxScaler or StandardScaler?
*If you have outliers or non-normal distributions, QuantileTransformer can make clustering or PCA behave better.
*It makes the features more evenly spread.

NOTE: Quantile transformation is most effective on larger datasets. With only 10 rows, the transformation will be very coarse — just 10 possible values per column — so visual differences might be hard to interpret. It’s still useful to learn how it works, though.

```python
# Import the transformer 
from sklearn.preprocessing import QuantileTransformer

# Initialize the parameters
my_quantile_scaler = QuantileTransformer(n_quantiles=min(100, len(df_audio_features)),
                                         random_state=42)

# Fit the transformer to the data 
my_quantile_scaler.fit_transform(df_audio_features)

# Transform the data using the transformer
quantile_audio_features = my_quantile_scaler.transform(df_audio_features)

# Convert transformed data into the dataframe
quantile_scaled_df = pd.DataFrame(quantile_audio_features, 
                                 columns = df_audio_features.columns,
                                 index = df_audio_features.index)

# Display 'energy' column of original dataframe and scaled dataframe
pd.DataFrame({'original_energy_column' : df_audio_features.iloc[:,1],
             'quantile_scaled_column' : quantile_scaled_df.iloc[:,1]}
             ).sort_values(
                          by='original_energy_column'
                          )
```

# Power Transformer 


```python
# Import the transformer
from sklearn.preprocessing import PowerTransformer

# Initialize the parameters
my_power_transformer = PowerTransformer()

# Fit the data to the transformer 
my_power_transformer.fit_transform(df_audio_features)

# Transform the data using the transformer
powert_audio_features = my_power_transformer.transform(df_audio_features)

# Convert the transformed data into a dataframe
pt_scaled_df = pd.DataFrame(powert_audio_features, 
                           columns = df_audio_features.columns,
                           index = df_audio_features.index)

# Display the original energy column and scaled column
pd.DataFrame({'original_energy_column' : df_audio_features.iloc[:, 1],
              'pt_scaled_column' : pt_scaled_df.iloc[:,1]}
             ).sort_values(
                          by='original_energy_column'
                          )
```
